# **Problem 1**
## **Problem 1.1**
### **Problem 1.1.1**
Our inequality: $C_1 \|x\|_{2} \leq \|x\|_{\infty} \leq C_2 \| x\|_{2}$

Vector norms are defined as:
* $\|x\|_{2} = \sqrt{\sum_{i=1}^{n} |x_{i}|^{2}}$
* $\|x\|_{\infty} = \max\limits_{1 \leq i \leq n}|x_{i}|$ 

1) Let's consider inequality $C_1 \|x\|_{2} \leq \|x\|_{\infty}$

Applied vector norms we get $C_1 \sqrt{\sum_{i=1}^{n} |x_{i}|^{2}} \leq \max\limits_{1 \leq i \leq n}|x_{i}|$

In order to define $C_1$, let vector $x$ consists of $\max\limits_{1 \leq i \leq n}|x_{i}|$, i.e. $x_{1} = x_{2} = \dotsc = x_{n} = max(x)$. In this case $\|x\|_{2} \rightarrow max$ and $\|x\|_{2} = \sqrt{n \cdot |x_{i}|^{2}} \implies C_1 \cdot \sqrt{n} \cdot |x_{i}| \leq |x_{i}| \implies C_1 \leq \frac{1}{\sqrt{n}}$

2) Let's consider inequality $\|x\|_{\infty} \leq C_2 \| x\|_{2}$

Applied vector norms we get $\max\limits_{1 \leq i \leq n}|x_{i}| \leq C_2 \sqrt{\sum_{i=1}^{n} |x_{i}|^{2}}$

In order to define $C_2$ we should minimize $\|x\|_{2}$, i.e. vector $x$ has to contain all zeros except one particular value $x_{i} \ne 0$. Then $\max\limits_{1 \leq i \leq n}|x_{i}| \leq C_2 \cdot |x_{i}|$. Ultimately, $\max\limits_{1 \leq i \leq n}|x_{i}| == |x_{i}| \implies C_2 \geq 1$

*Answer is*: $C_1 \leq \frac{1}{\sqrt{n}}, C_2 \geq 1$

### **Problem 1.1.2**
Need to prove $\| U A \|_F = \| A U \|_F = \| A \|_F$

$U$ is an unitary matrix. So, $$U^{*}U = UU^{*} = I$$, where $I$ is an identity matrix.

Frobenius norm is defined as $\|A\|_{F} = \sqrt{trace(A^{*}A)}$, where $trace(AB) = trace(BA)$.

1) $\|UA\|_{F} = \sqrt{trace((UA)^{*}UA)} = \sqrt{trace(A^{*}U^{*}UA)} = \sqrt{trace(A^{*}IA)} = \sqrt{trace(A^{*}A)} = \|A\|_{F}$

2) $\|AU\|_{F} = \sqrt{trace((AU)^{*}AU)} = \sqrt{trace(U^{*}A^{*}AU)}$


Let matrix $U^{*}A^{*}A = B$ and $U=C$, then $trace(BC) = trace(CB) \implies \|AU\|_{F} = \sqrt{trace(UU^{*}A^{*}A)} = \sqrt{trace(A^{*}A)} = \|A\|_{F}$

So, $\| U A \|_F = \| A U \|_F = \| A \|_F$. $Q.E.D$.

### **Problem 1.1.3**
Need to prove $\| U A \|_2 = \| A U \|_2 = \| A \|_2$

$2$-norm is defined as $\|A\|_{2} = \sqrt{\lambda_{max}(A^{*}A)}$

1) $\|UA\|_{2} = \sqrt{\lambda_{max}((UA)^{*}UA)} = \sqrt{\lambda_{max}(A^{*}U^{*}UA)} = \sqrt{\lambda_{max}(A^{*}IA)} = \sqrt{\lambda_{max}(A^{*}A)} = \|A\|_{2}$

2) Let's use definition of operator norms $$\|A\|_{p} = \sup_{x\ne0} \frac{\|Ax\|_{p}}{\|x\|_{p}}$$

$\|AU\|_{2} = \sup_{x\ne0} \frac{\|AUx\|_{2}}{\|x\|_{2}}$

Important mark: $\|Ux\|_{2} = \|x\|_{2}$

$\|AU\|_{2} = \sup_{x\ne0} \frac{\|AUx\|_{2}}{\|Ux\|_{2}}$, where product of $Ux$ is a vector. Thus, let's substitute our product $Ux$ to vector $y \implies \|AU\|_{2} = \sup_{x\ne0} \frac{\|Ay\|_{2}}{\|y\|_{2}} = \|A\|_{2}$. $Q.E.D$.

## **Problem 1.2**
### **Problem 1.2.1**

Need to prove $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$

Matrix Frobenius norm and 2-norm are defined as:
* $\|A\|_{2} = \delta_{max}(A)$
* $\|A\|_{F} = \sqrt{\sum_{i=1}^{m}\sum_{j=1}^{n}|a_{i,j}|^{2}}$

Let's describe singular value decomposition: $$A=U\Sigma V^{*}$$, where $U, V^{*}$ - unitary matrices, $\Sigma$ - diagonal matrix with singular values.

Important mark: if $rank(A) = r \Rightarrow \delta_{r+1} = \dotsc = \delta_{k} = 0$

Remark: All singular values in $\Sigma$ are sorted from $(1, 1)$ to $(n, n)$ in descending order.

Let $rank(A) = r$. Then it is possible to get the next inequality:

$\|U\Sigma V^{*}\|_{F} \leq \sqrt{r} \|A\|_{2} \Leftrightarrow \|\Sigma\|_{F} \leq \sqrt{r} \|A\|_{2}$

$\sqrt(r) \geq \frac{\|\Sigma\|_{F}}{\|A\|_{2}}$, where $\|\Sigma\|_{F}$, $\|A\|_{2} \geq 0$. Consequently, it is possible to square this inequality.

$r \geq \frac{\|\Sigma\|_{F}^{2}}{\|A\|_{2}^{2}}$, where $\|\Sigma\|_{F}^{2} = \sum_{i=1}^{m}\sum_{j=1}^{n}|a_{i,j}|^{2}$ - sum of singular values of the squared power and $\|\Sigma\|_{2}^{2} = \delta_{max}^{2}(A)$. Thus, it is able to represent our fraction in the form like: $$ r \geq \frac{x_{1,1}^{2} + x_{2, 2}^{2} + \dotsc x_{n, n}^{2}}{x_{1, 1}^{2}}$$, where $x_{1, 1}$ - maximal singular value, $n$ - number of singular values.

It is possible to represent $r = n \cdot \frac{x_{1,1}^{2}}{x_{1,1}^{2}}$. Thus, let's perform our inequality in the following way: $$n \cdot \frac{x_{1,1}^{2}}{x_{1,1}^{2}} \geq \frac{x_{1,1}^{2}}{x_{1,1}^{2}} + \frac{x_{2,2}^{2}}{x_{1,1}^{2}} + \dotsc + \frac{x_{n, n}^{2}}{x_{1,1}^{2}}$$, where $x_{1, 1} \geq x_{2, 2} \geq \dotsc \geq x_{n, n}$

It is obvious that left side is always greater or equal than right side, because $\frac{x_{2,2}^{2}}{x_{1,1}^{2}} \leq \dotsc \leq \frac{x_{n,n}^{2}}{x_{1,1}^{2}} \leq 1$. $Q.E.D$.

### **Problem 1.2.2**

a_{i,j} = \begin{cases} 1, i=j \leq k \\ 0, otherwise \end{cases}

### **Problem 1.2.3**
Need to prove if $\mathrm{rank}(A) = 1$, then $\| A \|_F  = \| A \|_2$.

Take advantage of definition of matrix norms from the __Problem 1.2.1__:
* $\|A\|_{2} = \delta_{max}(A)$
* $\|A\|_{F} = \sqrt{\sum_{i=1}^{m}\sum_{j=1}^{n}|a_{i,j}|^{2}}$

It is necessary to use singular value decomposition, which is also described in the __Problem 1.2.1__. After several transformations it is able to obtain the next equation: $\|A\|_{F} = \|U\Sigma V^{*}\|_{F} = \|\Sigma\|_{F}$, where $\Sigma$ is diagonal matrix with singular values in descending order.

$rank(A) = 1 \implies \delta_{2} = \delta_{3} = \dotsc = 0$. So, we have only one singular value, which is also a maximal singular value of matrix $A$.

In this way we can define $\|A\|_{F} = \|\Sigma\|_{F} = \sqrt{\sum_{i=1}^{m}\sum_{j=1}^{n}|a_{i,j}|^{2}} = |a_{1, 1}| = \delta_{max}(A) = \|A\|_{2}$. Q.E.D.

### **Problem 1.2.4**

Need to prove $\|AB\| \leq \|A\|_{2} \|B\|_{F}$

Singular value decomposition of matrix $A = U_{A}\Sigma_{A}V_{A}^{*}$

Then, $\|U_{A}\Sigma_{A}V_{A}^{*}B\|_{F} \leq \|A\|_{2}\|B\|_{F}$

After using preserving norm inequality has the form $\|\Sigma_{A}V_{A}^{*}B\|_{F} \leq \|A\|_{2}\|B\|_{F} \Leftrightarrow \|\Sigma_{A}V_{A}^{*}B\|_{F} \leq \|A\|_{2}\|V_{A}^{*}B\|_{F} $

Let $V_{A}^{*}B = M$, then $\|\Sigma_{A}M\|_{F} \leq \|A\|_{2}\|M\|_{F}$

Let's represent product of $\Sigma_{A}M$ in the form of element-wise product, i.e. $\|\Sigma_{A}M\|_{F} = \sqrt{\sum_{i=1}^{k}\sum_{j=1}^{n}|\delta_{i,j}m_{i,j}|^{2}}$, where $\|A\|_{2} = \delta_{max}(A) = \delta_{max}$ and $\|M\|_{F} = \sqrt{\sum_{i=1}^{k}\sum_{j=1}^{n}|m_{i,j}|^{2}}$

Then, inequality will take the following form: $\sqrt{\sum_{i=1}^{k}\sum_{j=1}^{n}|\delta_{i,j}m_{i,j}|^{2}} \leq \delta_{max} \cdot \sqrt{\sum_{i=1}^{k}\sum_{j=1}^{n}|m_{i,j}|^{2}}$

Left and right sides are greater or equal than zero, so it is possible to squared inequality. Then: $\sum_{i=1}^{k}\sum_{j=1}^{n}|\delta_{i,i}m_{i,j}|^{2} \leq \sum_{i=1}^{k}\sum_{j=1}^{n}\delta_{max}^{2} \cdot |m_{i,j}|^{2}$

It is obvious that the right side is always greater or equal than the left side, because $\delta_{max} > \delta_{2, 2} > \dotsc > \delta_{i,i}$.

So, $\|AB\| \leq \|A\|_{2} \|B\|_{F}. Q.E.D.$

````python
import jax 
import jax.numpy as jnp
import matplotlib.pyplot as plt
# import seaborn as sns
# sns.set()

%matplotlib inline
# plt.rc("text", usetex=True)
````

````python
@jax.jit
def F(X, U, V):
    frobenius = jnp.linalg.norm(X - U @ V, 'fro')
    return 1/2 * (frobenius ** 2)

gradient_U = jax.grad(F, argnums=1)
gradient_V = jax.grad(F, argnums=2)

def create_matrix(n_rows: int, n_columns: int):
    return jax.random.normal(jax.random.PRNGKey(0), (n_rows, n_columns))

@jax.jit
def analytical_grad_to_U(X, U, V):
    return (U @ V - X) @ V.T

@jax.jit
def analytical_grad_to_V(X, U, V):
    return U.T @ (U @ V - X)

@jax.jit
def automatic_grad_to_U(X, U, V):
    return gradient_U(X, U, V)

@jax.jit
def automatic_grad_to_V(X, U, V):
    return gradient_V(X, U, V)

n = 1000
k = 500

X = create_matrix(n, n)
U = create_matrix(n, k)
V = create_matrix(k, n)

print('\nAnalytical gradient of the function F  with respect to U')
%timeit analytical_grad_to_U(X, U, V).block_until_ready()

print('\nAutomatic gradient of the function F with respect to U')
%timeit automatic_grad_to_U(X, U, V).block_until_ready()

print('\nAnalytical gradient of the function F  with respect to V')
%timeit analytical_grad_to_U(X, U, V).block_until_ready()

print('\nAutomatic gradient of the function F with respect to V')
%timeit automatic_grad_to_U(X, U, V).block_until_ready()


# class MatrixCalculus:

#   def __init__(self, n: int, k: int):
#     self.X = self.create_matrix(n, n)
#     self.U = self.create_matrix(n, k)
#     self.V = self.create_matrix(k, n)

#   def create_matrix(self, n_rows: int, n_columns: int):
#     return jax.random.normal(jax.random.PRNGKey(0), (n_rows, n_columns))
  
#   @partial(jax.jit, static_argnums=(0,))
#   def analytical_grad_to_U(self):
#     return (self.U @ self.V - self.X) @ self.V.T

#   @partial(jax.jit, static_argnums=(0,))
#   def analytical_grad_to_V(self):
#     return self.U.T @ (self.U @ self.V - self.X)

#   @partial(jax.jit, static_argnums=(0,))
#   def automatic_grad_to_U(self):
#     return gradient_U(self.X, self.U, self.V)

#   @partial(jax.jit, static_argnums=(0,))
#   def automatic_grad_to_V(self):
#     return gradient_V(self.X, self.U, self.V)
````

````python
def calculate(n, k, key='analytical_U'):
    X = create_matrix(n, n)
    U = create_matrix(n, k)
    V = create_matrix(k, n)

    if key == 'analytical_U':
        t = %timeit -o -q analytical_grad_to_U(X, U, V).block_until_ready()

    elif key == 'automatic_U':
        t = %timeit -o -q automatic_grad_to_U(X, U, V).block_until_ready()

    elif key == 'analytical_V':
        t = %timeit -o -q analytical_grad_to_V(X, U, V).block_until_ready()

    elif key == 'automatic_V':
        t = %timeit -o -q automatic_grad_to_V(X, U, V).block_until_ready()

    return t.worst

def calculate_time(key, dimension, steps, const_value):
    if dimension == 'n':
        return [calculate(n=n, k=const_value, key=key) for n in steps]

    elif dimension == 'k':
        return [calculate(n=const_value, k=k, key=key) for k in steps]
````

````python
n_steps = jnp.linspace(600, 10000, 16, dtype='int32')
analytical_U_n = calculate_time('analytical_U', 'n', n_steps, 500)
automatic_U_n = calculate_time('automatic_U', 'n', n_steps, 500)
analytical_V_n = calculate_time('analytical_V', 'n', n_steps, 500)
automatic_V_n = calculate_time('automatic_V', 'n', n_steps, 500)
````

````python
fig = plt.figure(figsize=(22, 8))
ax = fig.add_subplot(111)    
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top=False, bottom=False, left=False, right=False)

ax1.scatter(n_steps, analytical_U_n, label='analytical - dF/dU')
ax1.scatter(n_steps, automatic_U_n, label='automatic - dF/dU')
ax2.scatter(n_steps, analytical_V_n, label='analytical - dF/dV')
ax2.scatter(n_steps, automatic_V_n, label='automatic - dF/dV')

ax.set_title(r'Dependence of worst running time per iteration and dimension N, where $k: const =500$', fontsize=18)
ax.set_xlabel('N dimension', fontsize=20)
ax.set_ylabel('Worst running time', fontsize=20)

ax.tick_params(axis="x", labelsize=20)
ax.tick_params(axis="y", labelsize=20)

ax1.legend(prop={'size': 16})
ax2.legend(prop={'size': 16})
````

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.fc2 = nn.Linear(1000, 1000)
        self.fc3 = nn.Linear(1000, 1000)
        self.fc4 = nn.Linear(1000, 1000)
        self.fc5 = nn.Linear(1000, 1000)
        self.fc6 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)

  
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

## Problem 3.2

````python
class AnalogueNet(nn.Module):

    def __init__(self, model, compression_rank):

        super(AnalogueNet, self).__init__()

        self.model = model
        self.r = compression_rank
        self.ReLU = self.model.ReLU

        self.fc2_layer = self.compress_layer(self.model.fc2)
        self.fc3_layer = self.compress_layer(self.model.fc3)
        self.fc4_layer = self.compress_layer(self.model.fc4)
        self.fc5_layer = self.compress_layer(self.model.fc5)

    def compress_layer(self, layer):
        layer_data = layer.weight.data
        u, s, v = torch.svd(layer_data, some=True)
        u, s, v = u[:, :self.r], s[:self.r], v[:, :self.r]

        layer_v = nn.Linear(1000, self.r, bias=False) #1000 x r
        layer_s = nn.Parameter(s) #r x r
        layer_u = nn.Linear(self.r, 1000, bias=True) #r x 1000

        with torch.no_grad():
            weights_v = torch.transpose(v, dim0=0, dim1=1)
            layer_v.weight.copy_(weights_v)

            layer_u.weight.copy_(u)
            layer_u.bias.copy_(layer.bias.data)

        return {'v': layer_v, 's': layer_s, 'u': layer_u}

    @staticmethod
    def linear_multiply(x, layer):
        layer_u = layer['u']
        layer_s = layer['s']
        layer_v = layer['v']

        x = layer_v(x) #Vx
        x = layer_u(layer_s * x) # USV_x
        return x

    def forward(self, x):

        #Input layer
        x = self.model.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)

        #FC2 layer
        x = self.linear_multiply(x, self.fc2_layer)
        x = self.ReLU(x)

        #FC3 layer
        x = self.linear_multiply(x, self.fc3_layer)
        x = self.ReLU(x)

        #FC4 layer
        x = self.linear_multiply(x, self.fc4_layer)
        x = self.ReLU(x)

        #FC5 layer
        x = self.linear_multiply(x, self.fc5_layer)
        x = self.ReLU(x)

        #Output layer
        x = self.model.fc6(x)
        return F.log_softmax(x, dim=1)

model1 = AnalogueNet(model, 350)
test(model1, test_loader)
````

````python
def get_accuracy(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))

    return correct / len(test_loader.dataset)
````

````python
def special_for_timeit(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    return correct / len(test_loader.dataset)
````

````python
import jax.numpy as jnp
import jax
import numpy as np
import random

class Game:
    def __init__(self, key):
        # key is a jax.random.PRNGKey
        self.key = key
        return

    def hermione_chooses_r_and_W(self):
        self.r = random.randint(1, 95)
        self.key, subkey = jax.random.split(self.key)
        self.W1 = jax.random.uniform(subkey, (self.r, 100), maxval=100., minval=0., dtype=jnp.float32)
        self.key, subkey = jax.random.split(self.key)
        self.W2 = jax.random.uniform(subkey, (100, self.r), maxval=100., minval=0., dtype=jnp.float32)

    def hermione_computes_function(self, x):
        return jnp.sin(self.W2 @ jnp.cos(self.W1 @ x))

    def harry_answers(self):
        n_rows = 100
        m = 100

        matrix = jax.random.normal(jax.random.PRNGKey(0), (n_rows, m))
        compute_scalar_func = lambda x: jax.random.choice(jax.random.PRNGKey(713),\
                                                    self.hermione_computes_function(x))
    
        gradient_f = jax.grad(compute_scalar_func, argnums=0)
        gradient_matrix = jnp.apply_along_axis(gradient_f, 1, matrix) / m ** 0.5
        u, s, v = jnp.linalg.svd(gradient_matrix)
        r = sum(1 if value // 1 else 0 for value in s)

        return r
    
    def play(self, n_rounds, verbose=True):
        # n_rounds: a number of rounds of the game
        # verbose: print or not the result of each round
        n_right_answers = 0
        for _ in range(n_rounds):
            self.hermione_chooses_r_and_W()
            r = self.harry_answers()
            if abs(r - self.r) == 0:
                if verbose:
                    print("Good job! The true answer is {}, Harry's answer is {}!".format(self.r, r))
                n_right_answers += 1
            else:
                if verbose:
                    print("Harry's answers is {}, but the true answer is {} :(".format(r, self.r))
        if float(n_right_answers) / n_rounds > 0.95:
            print('Well done: {}/{} right answers!'.format(n_right_answers, n_rounds))
        else:
            print('Only {}/{} right answers :( Work a little more and you will succeed!'.format(
                n_right_answers, n_rounds))
````
